```
def menu():
    print("\n=== MENU ===")
    print("1. A pangkat B")
    print("2. Hitung deret Fibonacci")
    print("0. Keluar")


def pangkat():
    try:
        angka = int(input("Masukkan suatu bilangan bulat: "))
        pangkat = int(input("Masukkan pangkat yang diinginkan: "))

        i = 1
        while i <= pangkat:
            print(f"{angka} pangkat {i} = {angka ** i}")
            i += 1

    except:
        print("Error: Mohon masukkan angka bulat saja.")


def deret_fibo():
    try:
        jumlah = int(input("Masukkan jumlah suku (N): "))
        total = 0

        deret = [1, 1]
        while len(deret) < jumlah * 2:
            deret.append(deret[-1] + deret[-2])

        index = 0
        for i in range(jumlah):
            atas = deret[index]
            bawah = deret[index + 1]
            nilai = atas / bawah

            if i % 2 == 0:
                total += nilai
                tanda = "+" if i != 0 else ""
            else:
                total -= nilai
                tanda = "-"

            print(f"{tanda} ({atas}/{bawah})", end=" ")
            index += 2

        print(f"\nHasil akhir: {total}")

    except:
        print("Error: Mohon masukkan angka bulat saja.")


while True:
    menu()
    pilih = input("Masukkan pilihan: ")

    if pilih == "1":
        pangkat()
    elif pilih == "2":
        deret_fibo()
    elif pilih == "0":
        print("Program selesai. Sampai jumpa!")
        break
    else:
        print("Pilihan tidak valid, silakan coba lagi.")
```
